```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("mitbih_train.csv", header=None)
df.head()
```

```python
conditions = {
    0: "Normal (N)",
    1: "Supraventricular premature (S)",
    2: "Premature ventricular contraction (V)",
    3: "Fusion of ventricular and normal beat (F)",
    4: "Unclassifiable beat (Q)"
}
```

```python
distribution = df[187].value_counts().to_dict()
distribution = {conditions[int(k)]: v for k, v in distribution.items()}

# Colores para el grÃ¡fico
palette = [
    "#264653",  # dark blue-green
    "#2a9d8f",  # teal
    "#1d3557",  # deep navy
    "#6d597a",  # muted purple
    "#b5838d"   # dusty rose
]

# Crear grÃ¡fico de pastel tipo "donut"
fig, ax = plt.subplots(figsize=(12, 6), facecolor='#f0f0f0')
wedges, texts, autotexts = ax.pie(
    list(distribution.values()), 
    labels=None, 
    colors=palette,
    pctdistance=0.85,
    autopct='%1.1f%%',
    textprops={'color': '#f0f0f0', 'fontweight': 'bold'},
    wedgeprops={'edgecolor': '#f0f0f0'}
)

# CÃ­rculo interno para el efecto donut
centre_circle = plt.Circle((0, 0), 0.70, fc='#f0f0f0')
ax.add_artist(centre_circle)

# Leyenda
ax.legend(
    wedges,
    list(distribution.keys()),
    title="Condiciones del corazon",
    loc="center left",
    bbox_to_anchor=(1, 0.5),
    fontsize=12
)

# TÃ­tulo y formato final
plt.title('Distribucion de categorias', fontsize=16, color='black')
plt.axis('equal')
plt.tight_layout()
plt.show()
```
